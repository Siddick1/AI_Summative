### Building models and prediction

#### Importing libraries for our machine learning model

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

**Plotting the data that we are going to use**

```python
sns.relplot(x='Relative management score',y='Country', data=data11)
```

***Training the data***

```python
train=data11.drop(['Country','Relative management score'], axis=1)
```

```python
test=data11['Relative management score']
```

```python
X_train,X_test,y_train, y_test=train_test_split(train,test,test_size=0.3,random_state=2)
```

***Linear Regression***

```python
regr=LinearRegression()
```

```python
regr.fit(X_train,y_train)
```

***Predicting***

```python
pred=regr.predict(X_test)
```

```python
pred
```

***Performance***

```python
regr.score(X_test,y_test)
```
